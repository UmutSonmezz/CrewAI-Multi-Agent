```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
os.environ["SERPER_API_KEY"] = ""
from dotenv import load_dotenv
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI
os.environ["GOOGLE_API_KEY"] = ""

llm=ChatGoogleGenerativeAI(model="gemini-1.5-flash",
                           verbose=True,
                           temperature=0.5,
                           google_api_key=os.getenv(""))


```

```python
search_tool = SerperDevTool()

researcher = Agent(
  role='Senior Research Analyst',
  goal='Find current solutions to use about {topic} error.',
  backstory="You have become adept at solving problems that arise."
            "Explains the solution to the problem and gives short answers",
  verbose=True,
  allow_delegation=False,
  llm=llm,
  max_iter =8,
  tools=[search_tool])

writer = Agent(
  role='Tech Content Strategist',
  goal=' Create a short answer that explains the solution clearly and concisely about {topic}.',
  backstory="You are a Content Strategist who explains the results of solutions in a descriptive and solution-oriented manner .",
  verbose=True,
  allow_delegation=False,
  llm=llm,
 )

solver = Agent(
  role='Instance Builder',
  goal='It takes the information created by the Tech Content Strategist and turns it into concrete examples, clarifying the solution path related to {topic}',
  backstory="A solution specialist who solves errors on websites in the most logical way by taking advantage of his previous experiences.",
  verbose=True,
  allow_delegation=False,
  llm=llm,)
```

```python
task1 = Task(
  description="""Do a detailed search on the solution to your {topic}.
  Research similar situations encountered and basic solutions.""",
  expected_output="Full analysis text with solutions",
  agent=researcher
)

task2 = Task(
  description="""
  Write an informative technical text about {topic}.
  The text should be informative, accessible and specific to those looking to solve website problems.
  This article should be easy to understand and solution oriented.""",
  expected_output="Full technical article about {topic} of no more than 3 paragraph ",
  agent=writer,
  #output_file='new-blog-post.md'
)

task3=Task(
  description="""
  "Take Information from Instance Builder.Give concrete suggestions on the {topic} solution.  "
 "These suggestions should be recommendations to solve errors on the website.",
 "It should be possible to implement on the website""",
  expected_output="The result should be short and concise",
  agent=solver)
```

```python
crew = Crew(
  agents=[researcher, writer,solver],
  tasks=[task1, task2, task3],
  verbose=2, # You can set it to 1 or 2 to different logging levels
  process = Process.sequential
)
result=crew.kickoff(inputs={'topic':"""I received an error message saying Background and foreground colors do not have a sufficient contrast ratio on a website, and the HTML element where the error occurs is
div.flex > div.flex > div.py-2 > button.grid. The problematic part is <<button type="button" class="grid place-content-center p-2 rounded-full hover:ring-2 mx-auto">>.Please resolve this error with code."""})
print(result)
```
